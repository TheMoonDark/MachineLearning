```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
#X = np.random.randn(100, 3)*10
n = 500
Y = np.random.randn(n, 3)
jishuzu = np.random.randn(300, 3)
plt.ion()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(Y[:, 0], Y[:, 1], Y[:, 2])
ax.set_xlim(-150, 150)
ax.set_ylim(-150, 150)
ax.set_zlim(-200, 200)
t = 0
while t <= 10:
  
    plt.pause(0.5)
    t += 1      

    for i in range(n):
        x = random.gauss(0, 20)
        y = random.gauss(0, 20)
        z = random.gauss(0, 20)
        Y[i, 0] += x
        Y[i, 1] += y
        Y[i, 2] += z
        if Y[i, 0] ** 2 + Y[i, 1] ** 2 > 100 ** 2:
            Y[i, 0] -= x
            Y[i, 1] -= y

        if Y[i, 2] > 150:
            Y[i, 2] -= z
        if Y[i, 2] < -150:
            Y[i, 2] += z
    sc._offsets3d = (Y[:, 0], Y[:, 1], Y[:, 2])
#     plt.ioff()
    plt.show()
```
